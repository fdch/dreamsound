```python
import glob
```

```python
filenames = [ f for f in sorted(glob.glob("./audio/*.wav"))]
```

```python
ds.load_audio(filenames)
```

```python
for waveform, name in zip(ds.audio[2:], filenames):
    ds.plot(d.dreamsound.np.log(d.dreamsound.librosa_mel(waveform)), label=name)
    d.dreamsound.plt.show()
    d.dreamsound.plt.close()
    d.dreamsound.ipd.display(d.dreamsound.ipd.Audio(waveform, rate=ds.sr))
```
